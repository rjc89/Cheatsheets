```python
import sqlite3
import pandas as pd

db = sqlite3.connect('/Users/constar/Documents/GitHub/Data_Science_Portfolio/SQL_cheatsheet/movie.db')

def run_query(query):
    return pd.read_sql_query(query,db)
```

## SELECT and LIMIT

```python
query = 'SELECT name, gender FROM actor LIMIT 5;'
run_query(query)
```

```python
query = 'SELECT * FROM actor LIMIT 5;'
run_query(query)
```

```python
query = 'SELECT * FROM movie LIMIT 5;'
run_query(query)
```

## SELECT DISTINCT

```python
query = '''
SELECT DISTINCT year
FROM movie
ORDER BY year ASC
LIMIT 5;
'''
run_query(query)
```

## ORDER BY ASC/DESC

```python
query = '''
SELECT year 
FROM movie
ORDER BY year ASC
LIMIT 5;
'''
run_query(query)
```

## WHERE

```python
query = '''
SELECT * 
FROM movie
WHERE year < 1900;
'''
run_query(query)
```

## LIKE (first letter or word queries) AND/OR 

```python
query = '''
SELECT * 
FROM movie
WHERE (title LIKE 'A%') AND (year < 1900)
ORDER BY year DESC
LIMIT 5;
'''
run_query(query)
```

```python
query = '''
SELECT * 
FROM movie
WHERE (title LIKE 'A%') OR (year < 1900)
ORDER BY year DESC
LIMIT 5;
'''

run_query(query)
```

```python
query = '''
SELECT * 
FROM movie
WHERE (title LIKE 'Car%')
ORDER BY year DESC
LIMIT 5;
'''

run_query(query)
```

```python
query = '''
SELECT * 
FROM movie
WHERE (title LIKE '% Car %')
ORDER BY year DESC
LIMIT 5;
'''

run_query(query)
```

## Aggregate Functions AVG, MIN, MAX, COUNT

```python
query = '''
SELECT MIN(year) AS "Earliest Movie"
FROM movie;
'''
run_query(query)
```

```python
query = '''
SELECT MAX(id)
FROM actor;
'''
run_query(query)
```

```python
query = '''
SELECT COUNT(*)
FROM movie
WHERE year < 1900;
'''
run_query(query)
```

```python
query = '''
SELECT COUNT(*) AS 'How many films from before 1900'
FROM movie
WHERE year < 1900;
'''

run_query(query)
```

```python
query = '''
SELECT AVG(year) AS "Average year"
FROM movie;
'''
run_query(query)
```

## GROUP BY

```python
query = '''
SELECT year as "Year", COUNT(*) AS "Number of Films"
FROM movie
GROUP BY year
ORDER BY COUNT(*) DESC
LIMIT 5;
'''

run_query(query)
```

## HAVING

```python
query = '''
SELECT name, COUNT(name) 
FROM role
GROUP BY name
HAVING COUNT(name) > 5
LIMIT 5;
'''
run_query(query)


# SELECT cust_country,COUNT(grade) 
# FROM customer 
# GROUP BY cust_country 
# HAVING COUNT(grade)>2;
```

## Arithmetic Operators

```python
query = '''
SELECT title, year
FROM movie
WHERE 2017 - year < 30
LIMIT 5;
'''
run_query(query)
```

## JOIN 

```python
%%html
<img src="Visual_SQL_JOINS_orig.jpg" width="700">
```

```python
query = '''
SELECT *
FROM movie
LEFT JOIN role
ON movie.id = role.movie_id
ORDER BY year ASC
LIMIT 10;
'''
run_query(query)
```

```python
query = '''
SELECT *
FROM role
INNER JOIN movie
ON movie.id = role.movie_id
ORDER BY year ASC
LIMIT 10;
'''
run_query(query)
```

```python
query = '''
SELECT *
FROM role 
LEFT JOIN movie
ON movie.id = role.movie_id
WHERE role.movie_id IS NULL
ORDER BY year ASC
LIMIT 10;
'''
run_query(query)
```
